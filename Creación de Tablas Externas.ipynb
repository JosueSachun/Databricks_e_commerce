```sql
USE e_commerce.bronze;
```

```sql
CREATE TABLE IF NOT EXISTS silver.customer(
  customer_id VARCHAR(50),
  customer_unique_id VARCHAR(50),
  zip_code_prefix VARCHAR(40),
  city VARCHAR(40),
  state VARCHAR(40),
  ingestion_timestamp TIMESTAMP,
  environment VARCHAR(50),
  file_date DATE,
  source_name STRING
)
USING DELTA
LOCATION 'abfss://silver@saexternaldatabricks.dfs.core.windows.net/customer'
```

```sql
CREATE TABLE IF NOT EXISTS silver.geolocation(
  zip_code_prefix VARCHAR(50),
  latitude DOUBLE,
  longitude DOUBLE,
  city VARCHAR(70),
  state VARCHAR(70),
  ingestion_timestamp TIMESTAMP,
  environment VARCHAR(50),
  file_date DATE,
  source_name STRING
)
USING DELTA
LOCATION 'abfss://silver@saexternaldatabricks.dfs.core.windows.net/geolocation'
```

```sql
CREATE TABLE IF NOT EXISTS silver.order_item(
  order_item_c_key VARCHAR(50),
  order_id VARCHAR(50),
  order_item_id VARCHAR(50),
  product_id VARCHAR(50),
  seller_id VARCHAR(50),
  shipping_limit_date TIMESTAMP,
  price DOUBLE,
  freight_value DOUBLE,
  ingestion_timestamp TIMESTAMP,
  environment VARCHAR(50),
  file_date DATE,
  source_name STRING
)
USING DELTA
LOCATION 'abfss://silver@saexternaldatabricks.dfs.core.windows.net/order_item'
```
