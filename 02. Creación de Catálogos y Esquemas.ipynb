##Crear Catálogos y Esquemas para el Proyecto
1. Catalog - e_commerce(Sin Managed Location)
2. Schemas - bronze, silver y gold(Con Managed Location)

```sql
CREATE CATALOG IF NOT EXISTS e_commerce;
```

```sql
USE CATALOG e_commerce;
```

```sql
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION "abfss://bronze@saexternaldatabricks.dfs.core.windows.net/";
```

```sql
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION "abfss://silver@saexternaldatabricks.dfs.core.windows.net/";
```

```sql
CREATE SCHEMA IF NOT EXISTS gold
MANAGED LOCATION "abfss://gold@saexternaldatabricks.dfs.core.windows.net/";
```
